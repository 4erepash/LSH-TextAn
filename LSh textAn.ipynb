```python
#откроем файлы и превратим их в датафреймы, чтобы было удобнее работать
import pandas as pd
data_dub1 = pd.read_csv('tweeter_dub1.csv', sep = ';')
data_dub2 = pd.read_csv('tweeter_dub2.csv', sep = ';')
```

Так как проверка не должна зависеть от формы слова и регистра, сначала все слова в тексте нужно привести к начальной форме и нижнему регистру. Воспользуемся для этого библиотекой Mystem. 

```python
#создадим список всех текстов
txts_dub1_l = data_dub1['текст поста'].tolist()
txts_dub2_l = data_dub2['текст поста'].tolist()
```

```python
txts_dub1_l
```

```python
from pymystem3 import Mystem
```

```python
#проходимся по списку текстов, сначала убирая знаки препинания, потому что они нам не нужны, потом лемматизируем и добавляем  в конечный список уже чистый текст
def lemmatize(list_of_texts):
    m = Mystem()
    final_list = []
    for text in list_of_texts:
        text = text.replace(',', '').replace('.', '').replace('!', '').replace('?', '')
        lemmas = m.lemmatize(text)
        final_list.append(''.join(lemmas))
    return final_list
```

```python
dub1_lemmatized_l = lemmatize(txts_dub1_l)
```

```python
dub2_lemmatized_l = lemmatize(txts_dub2_l)
```

Теперь добавим в наш датафрейм столбец с очищенными текстами.

```python
data_dub1['нормализованные тексты постов'] = dub1_lemmatized_l
data_dub2['нормализованные тексты постов'] = dub2_lemmatized_l
```

```python
data_dub1
```

```python
#теперь напишем функцию, которая будет считать тональность твита и выдаст нам список с тональностями для каждого твита.
def text_tones(tweets_list):
    positive_words = ['хороший', 'спасибо', 'радостный', 'вкусный', 'забавный', 'милый', 'уютный', 'веселый', 'мелодичный', 'свободный', 'ласковый', 'впечатляющий', 'талантливый']
    negative_words = ['ненавистный', 'безответственный', 'дикий', 'наказание', 'жестокость', 'убийство', 'беда', 'виновный', 'варварство', 'грустно', 'печально', 'плохой', 'фейк']
    tones_list = []
    for tweet in tweets_list:
        tone = 0
        for word in tweet.split():
            if word in positive_words:
                tone += 1
            elif word in negative_words:
                tone -= 1
        tones_list.append(tone)
    return tones_list
```

```python
data_dub1['тональность'] = text_tones(dub1_lemmatized_l)
data_dub2['тональность'] = text_tones(dub2_lemmatized_l)
```

```python
with open ('5letter_words.txt', encoding = 'utf-8') as f:
    data = f.readlines()
```

```python
words_wordle = [word[:-1] for word in data]
```

```python
import string
def input_check(users_input):
    for l in users_input:
        if l not in string.ascii_letters:
            print('ُِВведено не английское слово')
            return False
        else:
            pass
    if len(users_input) != 5:
        print('ُِВведено не пятибуквенное слово')
        return False
    else:
        pass
    if users_input not in words_wordle:
        print('ِВы ввели не существующее слово, попробуйте еще')
        return False
    else:
        return True
```

```python
def wordle_result(correct_word, users_word):
    correct_letters = set(correct_word)
    for l1,l2 in zip(users_word, correct_word):
        if l1 in correct_letters:
            if l1 == l2:
                print(l1, 'стоит на правильном месте')
            else:
                print(l1, 'есть, но стоит не там')
        else:
            print(l1,'нет в слове')
```

```python
import random
correct_word = random.choice(words_wordle)
n = 0
while n < 6:
    users_word = input("Введите пятибуквенное слово на английском языке ")
    while input_check(users_word) is False:
        users_word = input("Введите пятибуквенное слово на английском языке ")
    if users_word == correct_word:
        print('Поздравляем! Вы отгадали слово!')
        break
    else:
        wordle_result(correct_word, users_word)
        n += 1    
```
